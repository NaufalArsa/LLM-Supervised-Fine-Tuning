```python
# Load generated dataset
import pandas as pd

data = pd.read_csv('cleaned-dataset.csv')
data
```

```python
# Delete value if contain '-' so its like a missing value
data = data.replace('-', pd.NA)
```

```python
data.info()
```

```python
# Check missing values
data.isnull().sum()
```

```python
# Drop missing values in Jenjang Pendidikan
data = data.dropna(subset=['Jenjang Pendidikan'])
```

```python
# Drop if values in Jenjang Pendidikan are more than 3 characters
data = data[data['Jenjang Pendidikan'].str.len() <= 3]
```

```python
data.info()
```

```python
# Drop column Jurusan if contain missing values
data = data.dropna(subset=['Jurusan'])
```

```python
data.info()
```

```python
# Delete value if contain '-' so its like a missing value
data = data.replace(' -', pd.NA)
data = data.replace('.', pd.NA)
data = data.replace("'", pd.NA)
data = data.replace('undefined', pd.NA)
data = data.replace('none', pd.NA)
```

```python
# Missing value check
data.isnull().sum()
```

```python
# If Pekerjaan Posisi is missing, change Lama Bekerja to missing too
data.loc[data['Pekerjaan Posisi'].isnull(), 'Lama Bekerja'] = pd.NA
```

```python
data = data.replace(' - ', pd.NA)
data = data.replace(' . ', pd.NA)
data = data.replace(' -- ', pd.NA)
data = data.replace(' --- ', pd.NA)
data = data.replace('...', pd.NA)
data = data.replace(' ... ', pd.NA)
data = data.replace('(kosong)', pd.NA)
data = data.replace('_', pd.NA)
data = data.replace('ok', pd.NA)

# Change to missing if column Bidang Sertifikasi contains tahun/bulan
data.loc[data['Bidang Sertifikasi'].str.contains(r'\b(tahun|bulan|thn|bln)\b', na=False, case=False), 'Bidang Sertifikasi'] = pd.NA
```

```python
data = data.replace('---', pd.NA)
data = data.replace('--', pd.NA)
data = data.replace(' (kosong)', pd.NA)
data.loc[data['Sertifikasi'].str.contains(r'\b(tahun|bulan|thn|bln)\b', na=False, case=False), 'Sertifikasi'] = pd.NA
```

```python
# Delete values if Pekerjaan Posisi contain tahun/bulan
data.loc[data['Pekerjaan Posisi'].str.contains(r'\b(tahun|bulan|thn|bln)\b', na=False, case=False), 'Pekerjaan Posisi'] = pd.NA
```

```python
# Save cleaned dataset
data.to_csv('cleaned-dataset3.csv', index=False)
```
